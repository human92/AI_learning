```python
#使用パッケージのインポート
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
#シグモイド関数の定義
def sigmoid(x):
    e = math.e
    y = 1/(1+e**-x)
    return y.astype(np.float64)
```

```python
#ステップ関数の定義
def step(x):
    y = x > 0
    return y.astype(np.int)
```

```python
#使用する数列
x = np.arange(-10, 10, 0.1)
```

```python
#シグモイト関数の可視化
y = sigmoid(x)
plt.plot(x, y)
plt.show()
```

```python
#ステップ関数の可視化
y = step(x)
plt.plot(x, y)
plt.show()
```

```python
#シグモイド関数の定義
def sigmoid_other(x):
    return 1/(1+np.exp(-x))

#シグモイト関数の可視化
y = sigmoid_other(x)
plt.plot(x, y)
plt.show()
```

```python
#レル関数の定義
def Relu(x):
    return np.maximum(0, x)
```

```python
#レル関数の可視化
y = Relu(x)
plt.plot(x, y)
plt.show()
```
